# Luminosity

````python
from spice.models import IcosphereModel
from spice.spectrum import simulate_observed_flux
from spice.spectrum.spectrum import simulate_monochromatic_luminosity
import astropy.units as u
import jax.numpy as jnp
import numpy as np
from transformer_payne import Blackbody
import matplotlib.pyplot as plt
%matplotlib inline

from jax import config
config.update('jax_platform_name', 'cpu')
config.update('jax_enable_x64', True)
````

Let's use a simple, blackbody model just for the testing purposes, alogside with a solar-like model.

````python
bb = Blackbody()
model = IcosphereModel.construct(10000, 1., 1., [5777.], bb.parameter_names)
````

````python
# Read the BB_5777.flx file
bb_data = np.genfromtxt('/Users/mjablons/Documents/stellar-mesh-integration/tutorial/data/BB_5777.flx', delimiter=',')
bb_wavelengths = bb_data[:, 0]  # First column contains wavelengths
bb_fluxes = bb_data[:, 1]       # Second column contains fluxes
````

To test the bolometric luminosity value, let's choose a wide range of wavelengths.

````python
#vws = jnp.linspace(1., 100000., 100000)
vws = bb_wavelengths
````

The observed flux is calculated for the distance of 10 parsecs by default.

````python
observed_flux = simulate_observed_flux(bb.intensity, model, jnp.log10(bb_wavelengths))
````

````python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

# Plot model and reference in the first subplot
ax1.plot(vws, observed_flux[:, 0], color='black', label='SPICE')
ax1.plot(bb_wavelengths, bb_fluxes*5.08659e-10, color='red', label='IDL Result')
ax1.set_ylabel('Intensity [erg/s/cm$^3$]')
ax1.legend()

# Plot residuals in the second subplot
residuals = observed_flux[:, 0] - bb_fluxes*5.08659e-10
ax2.plot(vws, residuals, color='blue')
ax2.set_xlabel('Wavelength [$\\AA$]')
ax2.set_ylabel('Residuals [erg/s/cm$^3$]')
fig.suptitle('T=5777 K')

# Adjust layout for better spacing
plt.tight_layout()
````

# Filters

Let's use the ```Filter``` subclasses to check the passband luminosities.

````python
from spice.spectrum.filter import JohnsonCousinsB, JohnsonCousinsI, GaiaG, JohnsonCousinsV, GaiaRP, GaiaBP, TWOMASSJ, TWOMASSH, TWOMASSK

jc_b = JohnsonCousinsB()
jc_i = JohnsonCousinsI()
gaia_g = GaiaG()
gaia_rp = GaiaRP()
gaia_bp = GaiaBP()
jc_v = JohnsonCousinsV()

twomass_j = TWOMASSJ()
twomass_h = TWOMASSH()
twomass_k = TWOMASSK()
````

````python
jc_b.plot_filter_responses_for_wavelengths(vws, observed_flux[:, 0])
jc_i.plot_filter_responses_for_wavelengths(vws, observed_flux[:, 0])
gaia_g.plot_filter_responses_for_wavelengths(vws, observed_flux[:, 0])
jc_v.plot_filter_responses_for_wavelengths(vws, observed_flux[:, 0])
twomass_j.plot_filter_responses_for_wavelengths(vws, observed_flux[:, 0])
````

````python
jc_b.plot_filter_responses_for_wavelengths(bb_wavelengths, bb_fluxes*5.08659e-10)
jc_i.plot_filter_responses_for_wavelengths(bb_wavelengths, bb_fluxes*5.08659e-10)
gaia_g.plot_filter_responses_for_wavelengths(bb_wavelengths, bb_fluxes*5.08659e-10)
jc_v.plot_filter_responses_for_wavelengths(bb_wavelengths, bb_fluxes*5.08659e-10)
twomass_j.plot_filter_responses_for_wavelengths(bb_wavelengths, bb_fluxes*5.08659e-10)
````

# Bolometric luminosity of the Sun

The bolometric luminosity of the Sun is the energy that is outputted by the whole Sun. To calculate that, we need to obtain the total flux first.

````python
from spice.spectrum import luminosity, absolute_bol_luminosity
````

````python
solar_luminosity = luminosity(bb.flux, model, vws)
print("Sun's bolometric luminosity:", solar_luminosity, "erg/s")
print("Sun's absolute bolometric luminosity:", absolute_bol_luminosity(solar_luminosity), "mag")
````

Let's compare that to the theoretical luminosity:

$L=4\pi R^2 \sigma T^4$

````python
sigma = (5.67*1e-8*u.W/(u.m**2)/(u.K**4)).to(u.erg/(u.cm**2)/(u.s)/(u.K**4))

print("Theoretical luminosity of the Sun:", jnp.sum(model.areas)*(u.solRad.to(u.cm)**2)*sigma*(5772*u.K)**4, "erg/s")
````
