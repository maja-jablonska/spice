```python
from transformer_payne import TransformerPayne
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import jax.numpy as jnp
```

```python
tp = TransformerPayne.download()
```

```python
spec = tp.intensity(jnp.log10(jnp.linspace(3000, 10000, 1000)), 1.0, tp.solar_parameters)
```

```python
plt.plot(jnp.linspace(3000, 10000, 1000), spec[:, 0], color='black')
plt.xticks([])
plt.yticks([])
plt.xlabel('')
plt.ylabel('')
plt.title('')
```
